```python
# train_test_split - commonly used in machine learning to split data into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.20,random_state=0)

# test_size = 0.20    specifies the proportion of the dataset to include in the test split  20%[0.20] for test and  remaining 80% for train
```

```python

from sklearn.linear_model import LinearRegression   ## used to perform linear regression
regressor = LinearRegression()        #instance creates
regressor.fit(x_train,y_train)        #This line trains the linear regression model using the training data
```

```python
#  evaluate its performance on the test data
y_predict = regressor.predict(x_test)
print(y_predict)
```

```python
# various metrics (such as
# 1- Mean Absolute Error,
# 2- Mean Squared Error,
# 3- R-squared) to assess the performance of your model

from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_predict)    #y_test - actual target values from the test set
                                        #y_predict - predicted values generated by  model for the test set.
```
